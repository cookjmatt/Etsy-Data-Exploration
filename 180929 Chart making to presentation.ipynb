```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
import numpy as np
```

```python
x = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
y = [.03, .09, .18, .31, .53, .90, 1.35, 1.93, 2.38, 2.84, 3.25]

fig, ax = plt.subplots(figsize=(20,10))
plt.title("Etsy Total Annual Sales")
plt.xlabel("Year")
plt.xlim(2006.5, 2017.5)
plt.xticks([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
plt.ylabel("Billions $USD")
ax.bar(x, y, align='center')
rects = ax.patches
labels = y

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, label,
            ha='center', va='bottom')
plt.savefig('Etsy sales.png', bbox_inches='tight')
plt.show()
```
